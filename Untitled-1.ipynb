```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sd
```

### Load All The Raw DataSet 

```python
geo = pd.read_csv("C:/Users/arpit/OneDrive/Desktop/PostNL-Scissor-Effect-Analytics/PostNL-Raw-Data/PostNL_Dim_Geography.csv")
delivery = pd.read_csv("C:/Users/arpit/OneDrive/Desktop/PostNL-Scissor-Effect-Analytics/PostNL-Raw-Data/PostNL_Fact_Deliveries.csv")
labor = pd.read_csv("C:/Users/arpit/OneDrive/Desktop/PostNL-Scissor-Effect-Analytics/PostNL-Raw-Data/PostNL_Fact_Labor.csv")
```

### Add A New Column Calculation

```python
labor['Total_Labor_Cost_EUR']=round(labor['Hours_Worked']*labor['Hourly_Rate_EUR'],2)
```

```python
labor
```

### Use Agreegate & Groupby For A Better  View 

```python
labor1=labor.groupby(['Date','Depot_ID']).aggregate({'Total_Labor_Cost_EUR':'sum','Hours_Worked':'sum'}).reset_index()
```

```python
labor1
```

```python
delivery1=delivery.groupby(['Date','Depot_ID']).aggregate({'Parcel_ID':'count','Distance_km':'sum'}).reset_index()
```

```python
delivery1
```

```python
delivery1.rename(columns={'Parcel_ID':'Volume'},inplace=True)
```

```python
delivery1
```

### Join All the Table And convert In The One Single Table 

```python
cleaned_data=labor1.merge(delivery1,on=['Date','Depot_ID']).merge(geo,on=['Depot_ID'])
```

```python
cleaned_data
```

###  Calculate The Efficiency 

```python
cleaned_data['Unite_Per_Cost_EUR']=round(cleaned_data['Total_Labor_Cost_EUR']/cleaned_data['Volume'],2)
```

```python
cleaned_data
```

### Relation Between Unite_Per_Cost_EUR & Volume Of The Percels

```python
Correlation=cleaned_data['Unite_Per_Cost_EUR'].corr(cleaned_data['Volume'],method='spearman')
f'Corelation between Unite_Per_Cost_EUR and Volume is {Correlation:.4f}'
```

### Visualization With Seaborn And See The Scissor Effect

```python
sd.set_theme.style='dark_grey'
plt.figure(figsize=(15,6))
sd.lineplot(data=cleaned_data,x='Date',y='Unite_Per_Cost_EUR',label='Cost')
sd.lineplot(data=cleaned_data,x='Date',y='Volume',label='Volume')
plt.fill_between(cleaned_data['Date'],cleaned_data['Unite_Per_Cost_EUR'],cleaned_data['Volume'],color='red',alpha=0.1)
plt.title('PostNL Scissor Effect: Cost VS Volume',fontsize=18);
```
